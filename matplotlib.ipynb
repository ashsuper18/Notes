### Two Plot in One chart

```python
x=[1,2,3]
y=[4,5,6]
x2=[1,4,9]
y2=[6,7,8]
```

```python
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
from cProfile import label
from matplotlib import style
style.use('ggplot')
plt.plot(x, y, label= 'x and y',linewidth=3)
plt.plot(x2, y2, label='x2 and y2')
plt.legend()
plt.show()
```

```python
plt.scatter(x, y)
plt.scatter(x2, y2)
plt.show()
```
